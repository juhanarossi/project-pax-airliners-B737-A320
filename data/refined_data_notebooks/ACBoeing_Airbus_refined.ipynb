```python
import pandas as pd
import os
```

```python
df = pd.read_csv('types_and_annual_pax2.csv')
df.head()
```

```python
# Remove 'pax_' from the column names
new_column_names = {col: col.replace('pax_', '') for col in df.columns}

# Rename the columns in the DataFrame
df = df.rename(columns=new_column_names)
```

```python
df.head()
```

```python
# Filter rows where 'Description' column contains the string '737'
filtered_df = df[df['Description'].str.contains('737')]

# Create a new DataFrame with the filtered rows
boeing_df = filtered_df.copy()
```

```python
boeing_df
```

```python
# Reset the index and drop the old index
boeing_df = boeing_df.reset_index(drop=True)
```

```python
boeing_df.columns
```

```python

# Assuming you have a DataFrame named boeing_df
# First, create the new DataFrame BAcount_df with the specified columns
BAcount_df = boeing_df[['Code', 'Description', '2022']].copy()

# Next, drop rows with NaN values in the '2022' column
BAcount_df = BAcount_df.dropna(subset=['2022'])

# Calculate the sum of '2022' values
total_2022 = BAcount_df['2022'].sum()

# Calculate the percentage for each row's value in the '2022' column
BAcount_df['percentage'] = (BAcount_df['2022'] / total_2022) * 100

# Reset the index of the DataFrame
BAcount_df.reset_index(drop=True, inplace=True)

BAcount_df
```

```python
BAcount_df.to_excel('BAcount.xlsx', index=False)
```

```python
# Create a regex pattern to match any of the strings
pattern = r'A318|A319|A320|A321'

# Filter the DataFrame to get rows with the desired strings in the 'Description' column
A320_df = df[df['Description'].str.contains(pattern, regex=True)]

# Display the filtered DataFrame
A320_df
```

```python
A320_df = A320_df.drop('Code', axis=1)
A320_df
```

```python
A320_df = A320_df.reset_index(drop=True)
```

```python
A320count_df = A320_df[['Description', '2022']].copy()

# Next, drop rows with NaN values in the '2022' column
A320count_df = A320count_df.dropna(subset=['2022'])

# Calculate the sum of '2022' values
sum_2022 = A320count_df['2022'].sum()

# Calculate the percentage for each row's value in the '2022' column
A320count_df['percentage'] = (A320count_df['2022'] / sum_2022) * 100

# Reset the index of the DataFrame
A320count_df.reset_index(drop=True, inplace=True)

A320count_df
```

```python
A320count_df.to_excel('A320count.xlsx', index=False)
```

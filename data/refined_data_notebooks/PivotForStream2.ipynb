```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv('types_and_annual_pax2.csv')
df1
```

```python
df1.dtypes
```

```python
# Melt the DataFrame to create a long-format DataFrame
melted_df = pd.melt(df1, id_vars=['Code', 'Description'], var_name='year', value_name='pax_count')

# Convert 'year' column to datetime dtype
melted_df['year'] = pd.to_datetime(melted_df['year'], format='%Y').dt.year

# Sort the melted DataFrame by 'year' and 'pax_count'
melted_df = melted_df.sort_values(['year', 'pax_count'], ascending=[True, False])

# Group by 'year' and get the top 10 values for each year
top_10_by_year = melted_df.groupby('year').head(10)

# Reset the index
pivot_df = top_10_by_year.reset_index(drop=True)
```

```python
melted_df
```

```python
pivot_df.head(20)
```

```python
# Drop the 'Code' column
pivot_df = pivot_df.drop('Code', axis=1)

# Rename the 'Description' column to 'description'
pivot_df = pivot_df.rename(columns={'Description': 'description'})

# Rearrange the columns
pivot_df = pivot_df[['year', 'description', 'pax_count']]
```

```python
pivot_df.to_csv('top10nineties.csv', index=False)
```

```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv('ACTypes.csv')
df1.head()
```

```python

# Define a function to read a CSV file and join it with df1
def join_csv_with_df1(year, df1):
    # Read the CSV file
    csv_path = f'PaxData/{year}.csv'
    csv_df = pd.read_csv(csv_path)

    # Convert 'Code' and 'Type' columns to the same data type (e.g., string)
    df1['Code'] = df1['Code'].astype(str)
    csv_df['Type'] = csv_df['Type'].astype(str)

    # Join df1 with the CSV DataFrame on "Code" and "Type"
    merged_df = df1.merge(csv_df, left_on='Code', right_on='Type', how='left')

    return merged_df

# Initialize an empty DataFrame
merged_df = pd.DataFrame()

# Loop through the years from 1990 through 1999
for year in range(1990, 2023):
    # Call the function for each year and join the DataFrames
    year_df = join_csv_with_df1(year, df1)
    if merged_df.empty:
        merged_df = year_df
    else:
        # Merge the DataFrames horizontally
        merged_df = pd.concat([merged_df, year_df], axis=1)

# Remove duplicate "Code" and "Description" columns
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]

merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['Type'])

merged_df
```

```python
merged_df.to_csv('types_and_annual_pax2.csv', index=False)
```
